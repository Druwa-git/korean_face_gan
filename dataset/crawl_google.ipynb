```python
from urllib.request import urlopen
from urllib.request import urlretrieve
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import os

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
            print('created '+directory)
    except OSError:
        print('Error : already Created' + directory)

search = input('검색 ')
createFolder('data/'+search)
print('1')
url =f'https://www.google.com/search?q=%EC%97%B0%EC%98%88%EC%9D%B8+%ED%94%84%EB%A1%9C%ED%95%84+%EC%82%AC%EC%A7%84&tbm=isch&ved=2ahUKEwjaiPLu09ryAhUjJaYKHbtFC8EQ2-cCegQIABAA&oq=%EC%97%B0%EC%98%88%EC%9D%B8+%ED%94%84%EB%A1%9C%ED%95%84+%EC%82%AC%EC%A7%84&gs_lcp=CgNpbWcQAzIFCAAQgAQyBAgAEBgyBggAEAUQHjIGCAAQBRAeOggIABCABBCxAzoICAAQsQMQgwE6BAgAEAM6CwgAEIAEELEDEIMBOgYIABAIEB46BAgAEB5QmteCAViD9YIBYPT1ggFoCHAAeASAAYgCiAGrGpIBBjAuMjUuMZgBAKABAaoBC2d3cy13aXotaW1nsAEAwAEB&sclient=img&ei=Is8tYdrvD6PKmAW7i62IDA'
print("2")
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
print("3")
options = webdriver.ChromeOptions()
print("4")
options.add_experimental_option('excludeSwitches', ['enable-logging'])
print("5")
driver.get(url)
print("6")
for i in range(500):
    driver.execute_script("window.scrollBy(0,100000)") # 스크롤 범위
html = driver.page_source
print("7")
soup = BeautifulSoup(html, features="html.parser")
print("8")
img = soup.select('img')

n = 1
imgurl = []

for i in img:
    try:
        imgurl.append(i.attrs["src"])
    except KeyError:
        imgurl.append(i.attrs["data-src"])

for i in imgurl:
    urlretrieve(i, "data/"+ search +'/' + search + str(n) + ".jpg")
    n += 1
    print('downloading.........{}'.format(n))

driver.close()
```
